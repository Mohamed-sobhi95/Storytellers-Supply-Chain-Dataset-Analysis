```python
df = df.loc[~df.duplicated(subset=['coaster_name','Location','opening_date_clean'])].reset_index(drop=True).copy()
df.shape
```

```python
co=df['year_introduced'].value_counts()\
.head(10)\
.plot(kind='bar',title='Top 10 Years Coasters Introduced',color='red')
co.set_xlabel('Year Introduced')
co.set_ylabel('Count')
```

```python
df['speed_mph'].plot(kind='hist',title='Speed of Coasters',bins=20,color='red')
plt.xlabel('Speed (mph)')
```

```python
df['speed_mph'].plot(kind='kde',title='Speed of Coasters',color='blue')
plt.xlabel('Speed (mph)')
```

```python
df.plot(kind='scatter',x='speed_mph',y='height_ft',color='green',title='coaster speed vs. height')
plt.xlabel('Speed (mph)')
```

```python
sns.scatterplot(x='speed_mph',
                y='height_ft',
                hue='year_introduced',
                data=df)
```

```python
sns.pairplot(df,vars=['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce'],hue='Type_Main')
```

```python
df[['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce']].dropna().corr()
```

```python
sns.heatmap(df[['year_introduced','speed_mph','height_ft','Inversions_clean','Gforce']].dropna().corr(),annot=True)
```

```python
co=df.query('Location!="other"').groupby('Location')['speed_mph'].agg(['mean','count'])\
                        .sort_values('mean')\
                        .query('count >10')['mean']\
                        .plot(kind='barh',title='Average Coast Speed by Location',color='purple',figsize=(12,5))
co.set_xlabel('average coaster speed')
```
